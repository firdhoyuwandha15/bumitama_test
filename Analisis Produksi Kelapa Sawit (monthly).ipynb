```python
avg=np.zeros((14,12))
i=0
for column in dt_produksi:
    if column != 'Tanggal':
        for j in range(12):
            avg[i][j]=dt_produksi[column][j:120:12].mean()
        i+=1
```

```python
#Rata-rata curah hujan dan hari hujan berdasarkan bulan
ch=np.zeros(12)
ch[0]=dt_hujan['Curah Hujan'][0:120:12].mean()
ch[1]=dt_hujan['Curah Hujan'][1:120:12].mean()
ch[2]=dt_hujan['Curah Hujan'][2:120:12].mean()
ch[3]=dt_hujan['Curah Hujan'][3:120:12].mean()
ch[4]=dt_hujan['Curah Hujan'][4:120:12].mean()
ch[5]=dt_hujan['Curah Hujan'][5:120:12].mean()
ch[6]=dt_hujan['Curah Hujan'][6:120:12].mean()
ch[7]=dt_hujan['Curah Hujan'][7:120:12].mean()
ch[8]=dt_hujan['Curah Hujan'][8:120:12].mean()
ch[9]=dt_hujan['Curah Hujan'][9:120:12].mean()
ch[10]=dt_hujan['Curah Hujan'][10:120:12].mean()
ch[11]=dt_hujan['Curah Hujan'][11:120:12].mean()

hh=np.zeros(12)
hh[0]=dt_hujan['Hari Hujan'][0:120:12].mean()
hh[1]=dt_hujan['Hari Hujan'][1:120:12].mean()
hh[2]=dt_hujan['Hari Hujan'][2:120:12].mean()
hh[3]=dt_hujan['Hari Hujan'][3:120:12].mean()
hh[4]=dt_hujan['Hari Hujan'][4:120:12].mean()
hh[5]=dt_hujan['Hari Hujan'][5:120:12].mean()
hh[6]=dt_hujan['Hari Hujan'][6:120:12].mean()
hh[7]=dt_hujan['Hari Hujan'][7:120:12].mean()
hh[8]=dt_hujan['Hari Hujan'][8:120:12].mean()
hh[9]=dt_hujan['Hari Hujan'][9:120:12].mean()
hh[10]=dt_hujan['Hari Hujan'][10:120:12].mean()
hh[11]=dt_hujan['Hari Hujan'][11:120:12].mean()
```

```python
Bulan = ['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
data = {'Bulan': Bulan, 'Rata-Rata Produksi KS TT 1998': avg[0], 'Rata-Rata Produksi KS TT 2004': avg[3], 'Rata-Rata Produksi KS TT 2006': avg[5], 'Rata-Rata Curah Hujan (mm)': ch, 'Rata-Rata Hari Hujan (hari)': hh}
df = pd.DataFrame(data=data)
df
```

```python
fig, [ax1,ax2,ax3,ax4,ax5] = plt.subplots(5, 1, figsize=(10, 20))
ax1.plot(Bulan,avg1)
ax1.set_title('Rata-Rata Produksi berdasarkan Bulan untuk Tahun Tanam 1998',fontsize=20)

ax2.plot(Bulan,avg2)
ax2.set_title('Rata-Rata Produksi berdasarkan Bulan untuk Tahun Tanam 2004',fontsize=20)

ax3.plot(Bulan,avg3)
ax3.set_title('Rata-Rata Produksi berdasarkan Bulan untuk Tahun Tanam 2006',fontsize=20)

ax4.plot(Bulan,ch)
ax4.set_title('Rata-Rata Curah Hujan berdasarkan Bulan', fontsize=20)

ax5.plot(Bulan,hh)
ax5.set_title('Rata-Rata Hari Hujan berdasarkan Bulan', fontsize=20)

fig.tight_layout()

for x in range(12):
    ax1.axvline(x=x,color='red')
    ax2.axvline(x=x,color='red')
    ax3.axvline(x=x,color='red')
    ax4.axvline(x=x,color='red')
    ax5.axvline(x=x,color='red')
```

```python
#Correlation

i=0
for column in dt_produksi:
    if column!='Tanggal':
        c1=np.corrcoef(avg[i],ch)
        c2=np.corrcoef(avg[i],hh)
        print('Korelasi produksi sawit ', column, ' dengan (curah hujan, hari hujan): ',(c1[0,1],c2[0,1]))
        i+=1
```

```python
#Multiple Linear Regression
##Tahun tanam 1998
y = dt_produksi['Tahun Tanam 1998']
y = list(y)
dt_hujan0 = dt_hujan[24:144]
x1 = dt_hujan0[['Curah Hujan', 'Hari Hujan']]

x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
