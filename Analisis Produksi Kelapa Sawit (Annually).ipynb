```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
```

```python
#input data
dt_produksi = pd.DataFrame(pd.read_excel('Produksi Sawit (Annually).xlsx'))
dt_hujan = pd.DataFrame(pd.read_excel('Curah Hujan_Hari Hujan (Annually).xlsx'))
```

```python
#Visualisasi produksi sawit sesuai tahun tanam
fig, axs = plt.subplots(5, 3, figsize=(30, 30))
axs[0,0].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 1998'])
axs[0,0].set_title('Tahun Tanam 1998',fontsize=20)
axs[0,1].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 1999'])
axs[0,1].set_title('Tahun Tanam 1999',fontsize=20)
axs[0,2].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2003'])
axs[0,2].set_title('Tahun Tanam 2003',fontsize=20)
axs[1,0].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2004'])
axs[1,0].set_title('Tahun Tanam 2004',fontsize=20)
axs[1,1].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2005'])
axs[1,1].set_title('Tahun Tanam 2005',fontsize=20)
axs[1,2].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2006'])
axs[1,2].set_title('Tahun Tanam 2006',fontsize=20)
axs[2,0].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2007'])
axs[2,0].set_title('Tahun Tanam 2007',fontsize=20)
axs[2,1].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2008'])
axs[2,1].set_title('Tahun Tanam 2008',fontsize=20)
axs[2,2].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2009'])
axs[2,2].set_title('Tahun Tanam 2009',fontsize=20)
axs[3,0].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2010'])
axs[3,0].set_title('Tahun Tanam 2010',fontsize=20)
axs[3,1].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2011'])
axs[3,1].set_title('Tahun Tanam 2011',fontsize=20)
axs[3,2].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2012'])
axs[3,2].set_title('Tahun Tanam 2012',fontsize=20)
axs[4,0].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2013'])
axs[4,0].set_title('Tahun Tanam 2013',fontsize=20)
axs[4,1].plot(dt_produksi['Tahun'],dt_produksi['Tahun Tanam 2014'])
axs[4,1].set_title('Tahun Tanam 2014',fontsize=20)
```

```python
fig, [ax1,ax2] = plt.subplots(2, 1, figsize=(10, 10))
ax1.plot(dt_hujan['Tahun'],dt_hujan['Curah Hujan'])
ax1.set_title('Curah Hujan (mm)',fontsize=20)
ax2.plot(dt_hujan['Tahun'],dt_hujan['Hari Hujan'])
ax2.set_title('Hari Hujan (hari)',fontsize=20)
```

```python
#Correlation
##Hujan 0 tahun sebelum

i=0
for column in dt_produksi:
    if column!='Tahun':
        c1=np.corrcoef(dt_produksi[column],dt_hujan['Curah Hujan'][2:12])
        c2=np.corrcoef(dt_produksi[column],dt_hujan['Hari Hujan'][2:12])
        print('Korelasi produksi sawit ', column, ' dengan (curah hujan, hari hujan): ',(c1[0,1],c2[0,1]))
        i+=1
```

```python
#Correlation
##Hujan 1 tahun sebelum

i=0
for column in dt_produksi:
    if column!='Tahun':
        c1=np.corrcoef(dt_produksi[column],dt_hujan['Curah Hujan'][1:11])
        c2=np.corrcoef(dt_produksi[column],dt_hujan['Hari Hujan'][1:11])
        print('Korelasi produksi sawit ', column, ' dengan (curah hujan, hari hujan): ',(c1[0,1],c2[0,1]))
        i+=1
```

```python
#Correlation
##Hujan 2 tahun sebelum

i=0
for column in dt_produksi:
    if column!='Tahun':
        c1=np.corrcoef(dt_produksi[column],dt_hujan['Curah Hujan'][0:10])
        c2=np.corrcoef(dt_produksi[column],dt_hujan['Hari Hujan'][0:10])
        print('Korelasi produksi sawit ', column, ' dengan (curah hujan, hari hujan): ',(c1[0,1],c2[0,1]))
        i+=1
```

```python
#Multiple Linear Regression
#curah hujan + hari hujan 2 tahun sebelum
##Tahun tanam 1998
y = dt_produksi['Tahun Tanam 1998']
y = list(y)
dt_hujan0 = dt_hujan[0:10]
x1 = dt_hujan0[['Curah Hujan', 'Hari Hujan']]

x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
